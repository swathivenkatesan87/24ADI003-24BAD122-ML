SCENARIO 1: E-commerce Sales Data

Analyze a sample e-commerce dataset to understand product categories, sales trends, and missing information.
Dataset (Kaggle – Public): https://www.kaggle.com/datasets/carrie1/ecommerce-data


IN-LAB TASKS:

· Import necessary Python libraries.

· Load the dataset into a Pandas DataFrame.

· Inspect dataset using head(), tail(), info(), and describe().

· Check for missing values.

· Visualize sales per product using bar and line charts.

```python
# ===============================
# E-COMMERCE DATA ANALYSIS SCRIPT
# ===============================

# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
print("swathi-24bad122")
# 2. Load Dataset
# Change path to where you saved the CSV file
file_path = "data.csv.zip"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# 3. Inspect Dataset
print("\n--- FIRST 5 ROWS ---")
print(df.head())

print("\n--- LAST 5 ROWS ---")
print(df.tail())

print("\n--- DATA INFO ---")
print(df.info())

print("\n--- STATISTICAL SUMMARY ---")
print(df.describe())

# 4. Check Missing Values
print("\n--- MISSING VALUES ---")
print(df.isnull().sum())

# Optional: Drop missing values if needed
# df = df.dropna()

# 5. Sales per Product Calculation
# Assuming dataset columns: Quantity, UnitPrice, Description
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]

product_sales = df.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).head(10)

print("\n--- TOP 10 PRODUCTS BY SALES ---")
print(product_sales)

# 6. BAR CHART
plt.figure()
product_sales.plot(kind="bar")
plt.title("Top 10 Product Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()

# 7. LINE CHART
plt.figure()
product_sales.plot(kind="line", marker="o")
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

Import Libraries: It starts by importing pandas for data manipulation and matplotlib.pyplot for creating visualizations.
Load Dataset: It loads your e-commerce data from a CSV file named data.csv.zip into a pandas DataFrame called df. The encoding="ISO-8859-1" is specified to handle potential character encoding issues.
Inspect Dataset: This section provides initial insights into the data by printing:
The first and last 5 rows (df.head(), df.tail()) to get a quick look at the data structure.
A summary of the DataFrame including column names, non-null counts, and data types (df.info()).
Descriptive statistics for numerical columns, such as mean, standard deviation, min, max, and quartiles (df.describe()).
Check Missing Values: It identifies and displays the number of missing values in each column, which is crucial for data quality assessment.
Sales per Product Calculation: This is a key analytical step where:
A new column TotalPrice is calculated by multiplying Quantity and UnitPrice for each transaction.
It then groups the data by Description (product name), sums the TotalPrice for each product, sorts them from highest to lowest sales, and selects the top 10 products. This result is stored in the product_sales variable.
BAR CHART: It generates a bar chart visualizing the total sales for the top 10 products identified in the previous step. This helps in quickly comparing the sales performance of these products.
LINE CHART: Finally, it creates a line chart of the same top 10 product sales, which can sometimes provide a different perspective on the distribution or relative scale of sales among the top products, though for categorical data like product names, a bar chart is often more intuitive for comparison.


SCENARIO 2: Hospital Patient Records

Examine a patient dataset to identify missing health metrics and visualize distributions.

Dataset (Kaggle – Public): https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database


IN-LAB TASKS:

· Load the dataset into Pandas.

· Explore data structure and missing values.

· Visualize glucose levels and age distribution using histograms and boxplots.

· Identify patterns in health metrics. GIVE ME A CODE

```python
# ======================================
# HOSPITAL PATIENT RECORDS ANALYSIS
# Pima Indians Diabetes Dataset
# ======================================

# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load Dataset
# Change path to your downloaded CSV file
file_path = "diabetes.csv"
df = pd.read_csv(file_path)

# 3. Explore Dataset Structure
print("\n--- FIRST 5 ROWS ---")
print(df.head())

print("\n--- LAST 5 ROWS ---")
print(df.tail())

print("\n--- DATA INFO ---")
print(df.info())

print("\n--- STATISTICAL SUMMARY ---")
print(df.describe())

# 4. Check Missing Values
print("\n--- MISSING VALUES COUNT ---")
print(df.isnull().sum())

# In this dataset, zeros may represent missing values for some columns
cols_with_zero_missing = ["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]

df[cols_with_zero_missing] = df[cols_with_zero_missing].replace(0, pd.NA)

print("\n--- AFTER REPLACING ZERO WITH NA ---")
print(df.isnull().sum())

# 5. Visualize Glucose Distribution (Histogram)
plt.figure()
plt.hist(df["Glucose"].dropna(), bins=20)
plt.title("Glucose Level Distribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()

# 6. Visualize Age Distribution (Histogram)
plt.figure()
plt.hist(df["Age"].dropna(), bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 7. Boxplot for Glucose
plt.figure()
plt.boxplot(df["Glucose"].dropna())
plt.title("Glucose Boxplot")
plt.ylabel("Glucose Level")
plt.show()

# 8. Boxplot for Age
plt.figure()
plt.boxplot(df["Age"].dropna())
plt.title("Age Boxplot")
plt.ylabel("Age")
plt.show()

# 9. Identify Patterns (Correlation)
print("\n--- CORRELATION MATRIX ---")
print(df.corr(numeric_only=True))

# Optional Visualization of Correlation
plt.figure()
plt.imshow(df.corr(numeric_only=True))
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```

SCENARIO 3: Housing Dataset

Investigate a housing dataset for missing features and relationships between house attributes.

Dataset (Kaggle – Public): https://www.kaggle.com/datasets/harishkumardatalab/housing-price-prediction


IN-LAB TASKS:

· Load the dataset and inspect columns.

· Detect and summarize missing values.

· Visualize correlations between numerical features using scatter plots and heatmaps.

· Interpret trends for price prediction.

```python
# ======================================
# HOUSING DATA ANALYSIS
# ======================================

import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Dataset
df = pd.read_csv("Housing.csv")   # change path if needed

# 2. Inspect Columns
print("\n--- COLUMN NAMES ---")
print(df.columns)

print("\n--- DATA INFO ---")
print(df.info())

print("\n--- FIRST ROWS ---")
print(df.head())

# 3. Missing Values Summary
print("\n--- MISSING VALUES ---")
print(df.isnull().sum())

# 4. Correlation Matrix
corr = df.corr(numeric_only=True)
print("\n--- CORRELATION MATRIX ---")
print(corr)

# 5. Heatmap Visualization
plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.title("Feature Correlation Heatmap")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()

# 6. Scatter Plot (Area vs Price Example)
# Change column names if dataset differs
plt.figure()
plt.scatter(df["area"], df["price"])
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()

# 7. Scatter Plot (Bedrooms vs Price Example)
plt.figure()
plt.scatter(df["bedrooms"], df["price"])
plt.title("Bedrooms vs Price")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
```

SCENARIO 4: Banking Customer Data

Analyze banking data to detect missing account information and understand customer demographics.

Dataset (Kaggle – Public): https://www.kaggle.com/datasets/ajay1735/customer-personality-analysis

IN-LAB TASKS:

· Import dataset into Pandas.

· Examine structure, null values, and column types.

· Visualize age, income, and spending patterns using bar and box plots.

· Summarize insights from visualizations.

```python
# ======================================
# BANK CUSTOMER DATA ANALYSIS (CORRECT)
# ======================================

import pandas as pd
import matplotlib.pyplot as plt

# 1. LOAD DATASET (TAB SEPARATED FILE)
df = pd.read_csv("marketing_campaign.csv", sep="\t")

# 2. BASIC INSPECTION
print("\n--- FIRST 5 ROWS ---")
print(df.head())

print("\n--- DATA INFO ---")
print(df.info())

print("\n--- COLUMN NAMES ---")
print(df.columns)

# 3. CHECK MISSING VALUES
print("\n--- MISSING VALUES ---")
print(df.isnull().sum())

# 4. CREATE AGE COLUMN
df["Age"] = 2026 - df["Year_Birth"]

# 5. AGE GROUP BAR CHART
age_groups = pd.cut(df["Age"], bins=[0,25,35,45,55,65,100])
age_counts = age_groups.value_counts().sort_index()

plt.figure()
age_counts.plot(kind="bar")
plt.title("Customer Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. INCOME BOXPLOT
plt.figure()
plt.boxplot(df["Income"].dropna())
plt.title("Income Distribution")
plt.ylabel("Income")
plt.show()

# 7. TOTAL SPENDING COLUMN
df["TotalSpend"] = df[
    ["MntWines","MntFruits","MntMeatProducts",
     "MntFishProducts","MntSweetProducts","MntGoldProds"]
].sum(axis=1)

# 8. SPENDING BOXPLOT
plt.figure()
plt.boxplot(df["TotalSpend"])
plt.title("Customer Spending Distribution")
plt.ylabel("Total Spend")
plt.show()

# 9. SPENDING BY EDUCATION (BAR CHART)
edu_spend = df.groupby("Education")["TotalSpend"].mean()

plt.figure()
edu_spend.plot(kind="bar")
plt.title("Average Spending by Education")
plt.xlabel("Education Level")
plt.ylabel("Average Spend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 10. SUMMARY INSIGHTS
print("\n--- SUMMARY INSIGHTS ---")

print("Average Age:", df["Age"].mean())
print("Average Income:", df["Income"].mean())
print("Highest Spending Customer:", df["TotalSpend"].max())
print("Lowest Spending Customer:", df["TotalSpend"].min())
```
